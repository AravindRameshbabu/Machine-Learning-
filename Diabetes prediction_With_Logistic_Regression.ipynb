```python
y = data['Outcome']
```

```python
### Lets handle the imbalace data with imblearn
```

```python
import imblearn
```

```python
from imblearn.over_sampling import RandomOverSampler
random = RandomOverSampler()
```

```python
X_random_sample , y_random_sample = random.fit_resample(X,y)
```

```python
print(f'Shapeof_X_sample : ', X_random_sample.shape)
print(' '*20)
print('*'*20)
print(' '*20)
print(f'Shapeof_Y_Sample : ', y_random_sample.shape)
```

```python
### Lets check the outcome's value counts 

## After 

plt.figure(figsize=(12,10))
y_random_sample.value_counts().plot.pie(title='Outcome attribute after imblearn random_over_Sampler',autopct='%.2f' );
plt.show()
```

```python
## Before 

plt.figure(figsize=(12,10))
data['Outcome'].value_counts().plot.pie(title='Outcome attribute before handling the imbalance data',autopct='%.2f')
plt.show()
```

```python
## Lets split the data into train test split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_random_sample,y_random_sample,test_size=0.2,random_state=42)
```

```python
### Lets standarize the data with standard scaler
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
```

```python
X_train = std.fit_transform(X_train)
X_test = std.transform(X_test)
```

```python
### Lets print out the shape of our data 
```

```python
print(f'Shape of X_train : ',X_train.shape)
print(' '*20)
print('*'*20)
print(f'Shape of X_test : ',X_test.shape)
print(' '*20)
print('*'*20)
print(f'Shape of y_train :',y_train.shape)
print(' '*20)
print('*'*20)
print(f'Shape of y_test : ',y_test.shape)
```

```python
### Modelling the data 
```

## Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
Log = LogisticRegression()
```

```python
Log.fit(X_train,y_train)
```

#### Prediction Part

```python
Log_Test_Prediction = Log.predict(X_test)
Log_Train_Prediction = Log.predict(X_train)
```

#### Evaluating the Logistic regression's performance 

```python
from sklearn.metrics import accuracy_score,confusion_matrix,roc_auc_score,roc_curve,f1_score,classification_report
```

```python
print(f"Logistic_Regression's Accuracy_score_of_Test_prediction ",accuracy_score(Log_Test_Prediction,y_test)*100)
print(' '*20)
print('*'*20)
print(f"Logistic_Regression's Accuracy_score_of_Train_prediction ",accuracy_score(Log_Train_Prediction,y_train)*100)
print(' '*20)
print('*'*20)
print(f"Logistic_Regression's Confusion_Matrix_of_Test_prediction ",confusion_matrix(Log_Test_Prediction,y_test))
print(' '*20)
print('*'*20)
print(f"Logistic_Regression's Confusion_Matrix_of_Train_prediction",confusion_matrix(Log_Train_Prediction,y_train))
print(' '*20)
print('*'*20)
print(f"Logistic_Regression's F1 Score of Test_Prediction",f1_score(Log_Test_Prediction,y_test))
print(' '*20)
print('*'*20)
print(f"Logistic_Regression's F1 Score of Train_prediction",f1_score(Log_Train_Prediction,y_train))

print(' ' * 20 )
print(' ' * 20 )
print(' ' * 20 )
print(' ' * 20 )
print(' ' * 20 )

print('*'*20)

print(f"Logistic_Regression's Classification of Test_Prediction : "
      ,classification_report(Log_Test_Prediction,y_test))
print(' '*20)
print(' '*20)
print(' '*20)
print(' '*20)
print(' '*20)

print(f"Logistic_Regression's Classification of Train_Prediction : "
      ,classification_report(Log_Train_Prediction,y_train))
```

```python
## Better way to visualize the confusion matrix 
```

```python
cm_of_Test = confusion_matrix(Log_Test_Prediction,y_test)
cm_of_Train = confusion_matrix(Log_Train_Prediction,y_train)
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(cm_of_Test,annot=True,fmt='d',cmap='Blues',cbar=False,
           xticklabels=['Predicted 0' , 'Predicted 1'],
           yticklabels=['Actual 0 ', 'Actual 1'])
plt.xlabel('Prediction')
plt.ylabel('Actual')
plt.title('Confusion_Matrix')
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(cm_of_Train,annot=True,fmt='d',cmap='Greens',cbar=False,
           xticklabels=['Predicted 0 ', 'predicted 1'],
           yticklabels=['Actual 0 ', 'Actual 1'])

plt.xlabel('Prediction')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
